# Mengimport Liblary üóÉÔ∏è

```python
from google_play_scraper import app
import pandas as pd
import numpy as np
from google_play_scraper import  Sort, reviews
```

# Data Scrapping üìö

**Untuk Datasetnya, Mengambil semua ulasan dari aplikasi dengan ID com.shopee.id di Google Play Store.**

```python
result, continuation_token = reviews(
    'com.shopee.id',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=30000,
    filter_score_with=None
)
```

```python
shopee_df= pd.DataFrame(np.array(result),columns=['review'])
shopee_df = shopee_df.join(pd.DataFrame(shopee_df.pop('review').tolist()))
```

## Mengurutkan Dataset (Tanggal/at) **untuk melihat sajah**

```python
feature_df = shopee_df[['userName', 'score', 'at', 'content']]
feature_df = feature_df.sort_values(by='at', ascending=False)
feature_df.head()
```
